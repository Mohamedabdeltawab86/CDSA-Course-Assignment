```python
#import the main libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
sns.set(rc={'figure.figsize' : [9,9]}, font_scale=1.2)
```

```python
#loading the data
income = pd.read_csv('50lke.csv')
```

```python
#exploring the data, showing the upper 5 countries with largest GDP
income.head()
```

```python
sns.kdeplot(income['GDP'],  shade=True, color='m')
```

```python
income_corr = income.corr()
income_corr
```
